```python
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
#reading in the file to a pandas dataframe
sales = pd.read_excel(r'C:\Users\cnguyen\Desktop\RanchoPalosVerdes_Sales_2006-2020.xlsx')
```

```python
sales.head() #shows top 5 rows of dataframe. 
```

```python
#calculating percent of missing data 
percent_missing = sales.isnull().sum() * 100 / len(sales)
missing_value_df = pd.DataFrame({'column_name': sales.columns,
                                 'percent_missing': percent_missing})
missing_value_df.reset_index(drop = True, inplace = True)
missing_value_df.sort_values()
```

```python
#sorts the dataframe by percent_missing in ascending order
missing_value_df.sort_values('percent_missing', inplace = True)
```

```python
#grabbing the columns that have 50%+ of missing data into a list
removeColumns = missing_value_df.loc[missing_value_df.percent_missing > 50].column_name 
```

```python
#dropping the columns from the list above
sales.drop(removeColumns, axis = 1, inplace = True)
```
