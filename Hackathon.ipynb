```python
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
#reading in the file to a pandas dataframe
test = pd.read_excel(r'C:\Users\cnguyen\Desktop\RanchoPalosVerdes_Sales_2006-2020.xlsx')
```

```python
test.head() #shows top 5 rows of dataframe. 
```

```python
#calculating percent of missing data 
percent_missing = test.isnull().sum() * 100 / len(test)
missing_value_df = pd.DataFrame({'column_name': test.columns,
                                 'percent_missing': percent_missing})
```

```python
#sorts the dataframe by percent_missing in ascending order
missing_value_df.sort_values('percent_missing', inplace=True)
```
